# 2018 ~ 2022 충청북도 축제 정리

```python
cb1 = pd.read_excel('./data/festival/cb.xlsx', sheet_name='2018')
cb2 = pd.read_excel('./data/festival/cb.xlsx', sheet_name='2019')
cb3 = pd.read_excel('./data/festival/cb.xlsx', sheet_name='2020')
cb4 = pd.read_excel('./data/festival/cb.xlsx', sheet_name='2021')
```

```python
cb1.drop(columns=['연번','최초개최년도\n및 횟수'], inplace=True)
```

```python
cb2.drop(columns=['연번','최초개최년도\n및 횟수'], inplace=True)
cb3.drop(columns=['연번','최초개최년도\n및 횟수'], inplace=True)
cb4.drop(columns=['연번','최초개최년도\n및 횟수'], inplace=True)
```

```python
cb1.rename(columns={'2018년\n개최기간':'개최기간'}, inplace=True)
cb2.rename(columns={'2019년\n개최기간':'개최기간'}, inplace=True)
cb3.rename(columns={'2020년\n개최기간':'개최기간'}, inplace=True)
cb4.rename(columns={'2021년\n개최기간':'개최기간'}, inplace=True)
```

```python
real_cb = pd.concat([cb1,cb2,cb3,cb4], 
                    keys=['2018', '2019', '2020', '2021'])
```

```python
real_cb.dropna(inplace=True)
```
