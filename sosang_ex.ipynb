```python
# os 모듈 사용 예 
from glob import glob
from os.path import isdir

for x in glob('../*') :
    if isdir(x):
        print(x, '<DIR>')
    else :
        print(x)
```

```python
# 수집된 데이터의 모든 csv 파일을 읽어옴 : glob 사용
from glob import glob

file_names = glob('./sosang_store/*.csv')
file_names
```

```python
# 소상공인 상점관련 csv 파일들 합치기 
```

```python
# concat() 이용한 행결합 

total = pd.DataFrame()

for file_name in file_names :
    temp = pd.read_csv(file_name, encoding = 'utf-8')
    total = pd.concat([total, temp])
```

```python
total.info()
```

```python
total.head()
```

```python
total.reset_index(inplace=True, drop=True)
total
```

```python
total.size
total.shape
```

```python
total.isnull().sum()
```

```python
# 사용할 컬럼만 선택 
data = total[['상호명', '지점명', '상권업종대분류명', '상권업종중분류명', 
             '상권업종소분류명', '시도명', '시군구명', '행정동명',
             '경도', '위도']]
data
```

```python
#새로 생성한 데이터 프레임을 csv 파일로 저장
total.to_csv('sosang_store/total_store.csv', encoding = 'utf-8')
data.to_csv('sosang_store/data_store.csv', encoding = 'utf-8')
```
