```python
import folium
import pandas as pd
import json
```

# GeoJson을 활용한 경계 표현
- geojson : 행정구역과 같은 경계선이나 경로를 표현하는 데 효율적

```python
#경계선 관련 데이터 로딩
df_geo = pd.read_json('data/경기도행정구역경계.json')
df_geo
```

```python
with open('data/경기도행정구역경계.json') as f :
    gg_geo = json.loads(f.read())
```

```python
gg_geo
```

```python
#지도에 그리기
m = folium.Map([37, 127], 
              zoom_start = 8)

folium.GeoJson(gg_geo).add_to(m)
```

```python
m
```

```python
#구역 나눈 곳을 특정 데이터 값으로 표시 
# 경기도인구데이터활용
```

```python
gg_pp = pd.read_excel('data/경기도인구데이터.xlsx')
```

```python
gg_pp
```

```python
gg_pp.info()
```

```python
gg_pp.구분
```

```python
year = 2017

folium.Choropleth(geo_data = gg_geo,
                 data = gg_pp,
                 columns = ['구분', year],
                 key_on = 'feature.properties.name',
                  color = 'darkgreen',
                fill_color = 'YlGn',
                fill_opacity = 0.8,
                  line_opacity = 0.2,
                 legend_name = '인구수').add_to(m)
m
```

```python
#지도 저장
m.save('경기도' + str(year) + '인구수.html')
```
