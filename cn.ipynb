```python
cn2.drop(columns=['연번','최초개최년도\n및 횟수'], inplace=True)
cn3.drop(columns=['연번','최초개최년도\n및 횟수'], inplace=True)
cn4.drop(columns=['연번','최초개최년도\n및 횟수'], inplace=True)
```

```python
cn1.rename(columns={'2018년\n개최기간':'개최기간'}, inplace=True)
cn2.rename(columns={'2019년\n개최기간':'개최기간'}, inplace=True)
cn3.rename(columns={'2020년\n개최기간':'개최기간'}, inplace=True)
cn4.rename(columns={'2021년\n개최기간':'개최기간'}, inplace=True)
```

```python
real_cn = pd.concat([cn1,cn2,cn3,cn4], 
                    keys=['2018', '2019', '2020', '2021'])
```
